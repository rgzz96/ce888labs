## Task 1: Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import numpy as np
import seaborn as sns
```

```python
abs_df = pd.read_csv('2_Absenteeism_at_work.csv', sep=';')
audit_df = pd.read_csv('1_audit_risk.csv')
cancer_df = pd.read_csv('3_dataR2.csv')
```

```python
f, ax = plt.subplots(figsize=(10,6))
correlation = audit_df.corr()
heatm = sns.heatmap(round(correlation,2), annot=False, ax=ax, cmap="coolwarm", fmt='.2f', linewidths=.05)
f.subplots_adjust(top=0.93)
t = f.suptitle('Audit Dataset Correlation Heatmap', fontsize=14)
```

```python
f, ax = plt.subplots(figsize=(10,6))
correlation = abs_df.corr()
heatm = sns.heatmap(round(correlation,2), annot=False, ax=ax, cmap="coolwarm", fmt='.2f', linewidths=.05)
f.subplots_adjust(top=0.93)
t = f.suptitle('Absenteeism at work Dataset Correlation Heatmap', fontsize=14)
```

```python
f, ax = plt.subplots(figsize=(10,6))
correlation = cancer_df.corr()
heatm = sns.heatmap(round(correlation,2), annot=False, ax=ax, cmap="coolwarm", fmt='.2f', linewidths=.05)
f.subplots_adjust(top=0.93)
t = f.suptitle('Cancer Dataset Correlation Heatmap', fontsize=14)
```
